# Pre-Processing and Data Wrangling

# Train Test data split - To measure model evaluation in unbaised manner

```python
X = df.drop('isFraud', axis=1)
y = df['isFraud']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
X_train.info()
```

# Check for missing values

```python
X_train.isnull().sum()
```

# Eliminate automatilacally variables having more than 20% missing values

```python
X_train_before_filling = X_train
X_train = X_train[X_train.columns[X_train.isnull().mean() < 0.2]]
```

# Let us see which among remaining columns have missing values

```python
missing_cols = X_train.columns[X_train.isnull().mean() > 0]
missing_cols
```

```python
X_train['card5'].mean() # let us use this for comparison, keep it in mind
```

# Imputation: Filling missing value in a variable with reasonable approximation like mean of the variable to allow machine learning model works better

# 1. Single Imputation Technique

# Impute numeric variable with mean of the variable

Impute Numeric veriable with mean of the variable

```python
X_train_single = X_train
cols = X_train_single.columns
num_cols = X_train_single.select_dtypes(include=np.number).columns
X_train_single.loc[:, num_cols] = X_train_single.loc[:, num_cols].fillna(X_train_single.loc[:, num_cols].mean())
print(num_cols)
X_train_single.loc[:, num_cols].mean()
```

Impute catogorical variables with their mode

```python
cat_cols = list(set(cols) - set(num_cols))
X_train_single.loc[:, cat_cols] = X_train_single.loc[:, cat_cols].fillna(X_train_single.loc[:, cat_cols].mode().iloc[0])
print(cat_cols)
print(X_train_single.loc[:, cat_cols].mode().iloc[0])
```

# Check if Missing value is now 0 for all Variable

empty column list indicates that there is no missing value in variables

```python
X_train_single.columns[X_train_single.isnull().mean() > 0] # select variables with missing values
```

# Step#2: To make all veriable numeric to feed to machine learning model

```python
print(X_train.info())
X_train_dummy = pd.get_dummies(X_train, prefix_sep='-', drop_first=True)
```

```python
X_train_dummy
```

# Finalizing the data before training a model

```python
final_X_train = pd.DataFrame(data=X_train_dummy)
print(final_X_train.head())
print(final_X_train.shape)
```

```python
## Decision tree using grid search CV
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

params = {'max_depth': range(3, 20)}
clf = GridSearchCV(DecisionTreeClassifier(), params, n_jobs=-1, verbose=2, scoring= 'roc_auc')

clf.fit(final_X_train, y_train)
```

```python
dt = clf.best_estimator_ # final decision tree
print(clf.best_score_, clf.best_params_)
```

```python
import graphviz
```

```python
from sklearn.tree import export_graphviz
```

```python
dot_data = export_graphviz(dt, feature_names=final_X_train.columns,
                          class_names=['No Fraud', 'Fraud'],
                          filled=True, rounded=True,
                          special_characters=True)
graph = graphviz.Source(dot_data, format='png')
plt.figure(figsize=(16, 12))
graph
```

```python
from sklearn import tree
txt_data = tree.export_text(dt)
print(txt_data)
```

```python
plt.figure(figsize=(25, 20))
tree.plot_tree(dt, feature_names=final_X_train.columns,
                          class_names=['No Fraud', 'Fraud'],
                          filled=True)
```

# Handling Class Imbalance

# Using SMOTE

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter

over = SMOTE(sampling_strategy=0.2, random_state=2)
steps = [('o', over)]
pipeline = Pipeline(steps=steps)

X_res, y_res = pipeline.fit_resample(X_train_dummy, y_train)

print('Original dataset shape:', Counter(y_train))
print('Resampled dataset shape:', Counter(y_res))
```
