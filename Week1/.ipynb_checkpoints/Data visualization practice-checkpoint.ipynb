# Importing Packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

# Loading Dataset

```python
x = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv") #let's load a dataset
```

```python
x.head()
```

# Scatter Plot

* Scatter plot by matplotlib

```python
plt.scatter(x.crime_rate, x.percent_senior)
plt.show()
```

```python
plt.figure(figsize=(12, 8)) # Bubble plot
plt.scatter(x.crime_rate, x.percent_senior, s = np.arange(len(x.index))*10)
plt.title('Plot of Crime Rate vs Percent Senior')
plt.xlabel('Crime Rate')
plt.ylabel('Percent Senior');
```

* By Seaborn

```python
sns.scatterplot(x.crime_rate, x.percent_senior);
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='crime_rate', y='percent_senior',data=x, size=np.arange(len(x.index))*10)
plt.title('Plot of Crime Rate vs Percent Senior');
```

* By Pandas

```python
x.plot.scatter('crime_rate', 'percent_senior');
```

```python
x.plot.scatter('crime_rate', 'percent_senior', s=np.arange(len(x.index))*10, figsize=(12, 8))
plt.title('Plot of Crime Rate vs Percent Senior')
plt.xlabel('Crime Rate')
plt.ylabel('Percent Senior');
```

# Line Chart

```python
plt.figure(figsize=(12, 8))
plt.plot(x.work_force, x.income) # 2 arguments: X and Y points
plt.xlabel("Work Force") # Adding the label for the horizontal axis
plt.ylabel("Income")
plt.show()
```

```python
# formating the style
plt.figure(figsize=(12, 8))
plt.plot(x.work_force, x.income, 'r--o') # 2 arguments: X and Y points
plt.xlabel("Work Force") # Adding the label for the horizontal axis
plt.ylabel("Income")
plt.show()
```

```python
# formating the style
plt.figure(figsize=(12, 8))
plt.plot(x.work_force, x.income, 'ro') # 2 arguments: X and Y points
plt.xlabel("Work Force") # Adding the label for the horizontal axis
plt.ylabel("Income")
plt.show()
```

```python
# formating the style
plt.figure(figsize=(12, 8))
plt.plot(x.work_force, x.income, 'g+') # 2 arguments: X and Y points
plt.xlabel("Work Force") # Adding the label for the horizontal axis
plt.ylabel("Income")
plt.show()
```

```python
# adding consective plots
plt.plot(x.work_force, x.income, color="r") 
plt.show()
plt.plot(x.physicians, x.income) 
plt.show()
```

```python
plt.plot(x.work_force, x.income, color="r", label = 'work_force') 
plt.plot(x.physicians, x.income, label='physicians') 

# Adding a legend
plt.legend()

plt.show()
```

One row two columns

```python
plt.subplot(121) # row, column, index
plt.plot(x.work_force, x.income, 'go')
plt.title('Income vs Work Force')

plt.subplot(122).label_outer()
plt.plot(x.hospital_beds, x.income, "r^")
plt.title("Income vs Hospital Beds")

plt.suptitle('Sub Plots');
```

Two rows One column`

```python
plt.subplot(2,1,1) # row, column, index
plt.plot(x.work_force, x.income, "go")

plt.subplot(2,1,2)# row, column, index
plt.plot(x.hospital_beds, x.income, "r^")

plt.suptitle("Sub Plots")
plt.show()
```

two rows two columns

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))
ax[0,0].plot(x.work_force, x.income, 'go')
ax[0,1].plot(x.work_force, x.income, 'bo')
ax[1,0].plot(x.work_force, x.income, 'ro')
ax[1,1].plot(x.work_force, x.income, 'yo');
```

* By Seborn

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x.work_force, x.income, color='r') # 2 arguments: X and Y points
plt.xlabel("Work Force") # Adding the label for the horizontal axis
plt.ylabel("Income")
plt.show()
```

* Using Pandas

```python
x.plot.line('work_force', 'income', figsize=(12, 8), legend=False, linestyle='--', marker='o',color='r');
```
